### Chargement et nettoyage du fichier Transactions

```python
# Analyse transactions 679532 lignes, pas de val nul, 183 lignes en double ?, sur la col date 679371 val uniques, valeurs test_
transactions = pd.read_csv('transactions.csv')
display(transactions.head(3))
print(transactions.info(), '\n')

# Typage
# pd.to_datetime -> error
# je vais regarder les erreurs et je me rends compte quil s'agit de ligne avec la mention test
# je mets de côté et j'analyse en detail ces lignes ct_0 ct_1

print('La description du dataframe : \n', transactions.describe(), '\n')  
print('Les lignes en double ? : ', transactions.duplicated(keep=False).sum(), '\n')
print('Affichage du nombres de valeurs nulles sur chaque colonne : \n', transactions.isnull().sum(), '\n')
print('Le nombre de valeurs unique sur chaque colonne : \n', transactions.nunique(), '\n')
print('Vérification de la clé primaire du dataframe ?: \n', transactions.nunique() == transactions.shape[0], '\n')
```

```python
# L'analyse des lignes en double
mask = transactions.duplicated()
print(transactions.loc[mask])
print(transactions.loc[mask, ['id_prod', 'session_id', 'client_id']].value_counts())
# Creation d'un nouveau dataframe avec les lignes test pour les analyser
df_test = transactions.loc[transactions['date'].str.startswith('test')]
df_test['id_prod'].value_counts()
df_test['client_id'].value_counts()
print("L'impression du nouveau dataframe avec les enregistrements test :\n", df_test.head())

# Effacement des lignes test dans le df transactions
transactions.drop(transactions[transactions['date'].str.startswith('test')].index, inplace=True)
```

### Chargement et nettoyage du fichier Products

```python
# Analyse fichier products 3287 valeurs, pas de val null, pas de val en dulicates, clé primaire id_prod
products = pd.read_csv('products.csv')
display('Affichage de 3 premiers lignes de la dataframe : ', products.head(3))
print(products.info(), '\n')
print('La description du dataframe : \n', products.describe(include='all'), '\n')
print('Les lignes en double ? : ', products.duplicated(keep=False).sum(), '\n')
print('Affichage du nombre de valeurs unique sur chaque colonne : \n ', products.nunique(), '\n')
print('Vérification de la clé primaire du dataframe ?: \n', products.nunique() == products.shape[0], '\n')
print('Des lignes en double ? : ', products.isna().sum(axis=1).any(), '\n')
print('Les valeurs unique sur la col categ : \n ', products['categ'].unique())
# Recherche des produits avec le prix -1
print("\nRecherche des produits avec des prix negatifs")
display(products.loc[products['price'] <= 0, :])
# Recherche des id_prod T_0 qui correspond aux lignes de test
products.query('id_prod =="T_0"')
# Efacement de la ligne avec id_prod = T_0
products.drop(products[products['id_prod'] == 'T_0'].index, inplace = True)
print(products.info())
```

#### Analyse clients

```python
# la courbe de Lorenz des clients et du chiffre d'affaire : 4 clients sur 8600 realise 7% du Ciffre d'Affaire
tcp_groupe_client = tcp.groupby('client_id').sum('price')

# Le calcul de la courbe de Lorenz
ca = tcp_groupe_client[tcp_groupe_client['price'] > 0]
n=len(ca['price'])
lorenz = np.cumsum(np.sort(ca['price'])/ca['price'].sum())
lorenz = np.append([0], lorenz)
xaxis = np.linspace(0-1/n, 1+1/n, len(lorenz))
plt.plot(xaxis, lorenz, drawstyle='steps-post')
plt.plot([0,1], [0,1])
plt.annotate("4 clients réalisent 7% du CA", xy=(1, 0.93), xytext=(0.50, 0.95),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=10, color='black')
plt.xlabel("Pourcentage du nombre de clients", fontname='Gill Sans MT', fontsize=12)
plt.ylabel("Ventes", fontname='Gill Sans MT', fontsize=12)
plt.xticks(np.arange(0, 1.1, 0.1), [f"{int(val*100)}%" for val in np.arange(0, 1.1, 0.1)])
plt.yticks(np.arange(0, 1.1, 0.1), [f"{int(val*100)}%" for val in np.arange(0, 1.1, 0.1)])
plt.title("La courbe de Lorenz", fontname='Gill Sans MT', fontsize=14)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.show()

# calcul de gini 
AUC = (lorenz.sum() - lorenz[-1]/2 - lorenz[0]/2)/n
S = 0.5 - AUC
gini = 2*S
print("L'indicateur Gini est de : ", round(gini, 2))
```

### Il y a t-il un lien entre le genre du client et la categorie
H0 : Il n'y a pas de lien entre le genre du client et la catégorie.

HA : Il existe un lien significatif entre le genre du client et la catégorie.

```python
# réalisation de Heatmap des correlations entre les variables 
df_age['groupe_age'] = df_age['groupe_age'].astype('category')
df_age['sex'] = df_age['sex'].astype('category')

df_age['groupe_age_code'] = df_age['groupe_age'].cat.codes
df_age['sex_code'] = df_age['sex'].cat.codes

# Calculer la matrice de corrélation
correlation_matrix = df_age[['groupe_age_code', 'sex_code', 'age', 'categ', 'price']].corr()

# Créer un heatmap de la matrice de corrélation
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True,  center=0,
            xticklabels=['groupe_age_code', 'sex_code', 'age', 'categ', 'price'],
            yticklabels=['groupe_age_code', 'sex_code', 'age', 'categ', 'price'])
plt.title("Heatmap des corrélations entre variables")
plt.show()

# j'efface les deux colonnes ajoutés :groupe_age_code et sex_code
col_a_effacer = ['groupe_age_code', 'sex_code']
df_age = df_age.drop(col_a_effacer, axis=1)
# df_age.head()
```

```python
# diagrame a barres de la variable categ par sexe
print("Le diagrame a barres de la variable categorie par sexe")
sns.countplot(x='categ', hue='sex', data=df_age)
plt.style.use('seaborn')
plt.show()
alpha = 0.05
# Création du tableau de contingence
tableau_contingence = pd.crosstab(df_age['sex'], df_age['categ'])

# Effectuer le test du chi-squared
chi2, p, dof, expected = stats.chi2_contingency(tableau_contingence, correction=True)

# Calcul de la valeur du V de Cramer
n = tableau_contingence.sum().sum()
min_dim = min(tableau_contingence.shape) - 1
v_cramer = np.sqrt(chi2 / (n * min_dim))

# Affichage des résultats
print("Le tableau de contingence :")
print(tableau_contingence)
print("\nRésultats du test du chi-squared :")
print(f"Chi-squared = {chi2}")
print(f"P-value = {p}")
print("\nRésultats du test V de Cramer :")
print(f"V de Cramer = {v_cramer}")
if p < alpha:
    print("La p_value est inférieure à alpha, donc l'hypothèse nulle est rejetée.")
    print("Il existe un lien significatif entre le genre du client et la catégorie, mais la valeur de V de Cramer \nest très base ce qui nous indique que les deux variables ont très peu d'effet l'une sur l'autre. ")
else:
    print("La p_value est supérieure à alpha, donc il n'y a pas suffisamment de preuves pour rejeter l'hypothèse nulle.")
```

```python
# le lien entre l'age et la frequence d'achat 
alpha = 0.05
frec = df_age['age'].value_counts()
modalites = frec.index
df = pd.DataFrame(modalites, columns=['age'])
df['n'] = frec.values
df['f'] = df['n']/len(df_age)
df= df.sort_values('age')
df['F'] = df['f'].cumsum() 
display(df.head())

print("Graphique de la distribituion de la frequence d'achat par age")
plt.figure(figsize=(15,7))
sns.barplot(data=df, x='age', y='f')
plt.style.use('seaborn')
plt.xticks(rotation=90)
plt.show()

# Test de Shapiro-Wilk pour tester la normalité de l'age et de la frequence pour les petits échantillons
alpha=0.05
statistic, p_value = shapiro(df['age'])
print("Test de Shapiro-Wilk pour le groupe d'age:")
if p_value < alpha :
    print("p-value :", p_value, "\nLes données ne suivent pas une distribution normale\n ")
else :
    print("p-value :", p_value, "\nLes données suivent une distribution normale \n")


statistic, p_value = shapiro(df['f'])
print("Test de Shapiro-Wilk pour le groupe des frequences:")
if p_value < alpha :
    print("p-value :", p_value, "\nLes données ne suivent pas une distribution normale \n")
else :
    print("p-value :", p_value, "\nLes données suivent une distribution normale \n")
    

    
# Calcule des variances

# Le test de Levene de la variance 
statistic, p_value = levene(df['age'], df['f'])

print("Résultats du test de Levene :")
print(f"Statistique de test : {statistic}")
print(f"P-value : {p_value}")

if p_value < 0.05:
    print("Les variances ne sont pas égales.\n")
else:
    print("Les variances sont égales.\n")

    
# le test de Kruskal-Wallis
from scipy.stats import kruskal

# Effectuer le test de Kruskal-Wallis
statistic, p_value = kruskal(df['age'], df['f'])
print("Test de Kruskal-Wallis:")
print("Statistique de test :", statistic)
print("p-value :", p_value)
print(p_value < alpha)

print("La p-value est inferieure au niveau de signification de 0.05, \nCela suggère qu'il y a suffisamment de preuves pour rejeter l'hypothèse nulle. \nJe peux conclure qu'il y a un lien entre l'age des clients et la frequence d'achat.")
```

### Il y a t-il un lien entre l'age et le panier moyen

```python
# regroupement par session et age
df_panier_age = df_age.groupby(['session_id','age'])['price'].sum().reset_index()

# Le df de l'age et du Panier moyen par âge
df_panier_moyen = df_panier_age.groupby(['age'])['price'].mean().round(3).reset_index()

print("Graphique de la distribituion du panier moyen par age")
plt.figure(figsize=(15,7))
sns.barplot(data=df_panier_moyen, x='age', y='price')
plt.style.use('seaborn')
plt.xticks(rotation=90)
plt.show()


# Définition des seuils d'âge pour chaque catégorie
age_seuils = [31, 51, 94]

# Création d'une nouvelle colonne pour les catégories d'âge
df_panier_moyen['age_category'] = pd.cut(df_panier_moyen['age'], bins=[18] + age_seuils, labels=['19-31', '32-51', '52-94'])

# Création des listes de prix pour chaque catégorie d'âge
data_19_31 = df_panier_moyen[df_panier_moyen['age_category'] == '19-31']['price']
data_32_51 = df_panier_moyen[df_panier_moyen['age_category'] == '32-51']['price']
data_52_94 = df_panier_moyen[df_panier_moyen['age_category'] == '52-94']['price']

# Réalisation du test Shapiro-Wilk pour chaque catégorie d'âge
for category in df_panier_moyen['age_category'].unique():
    data = df_panier_moyen[df_panier_moyen['age_category'] == category]['price']
    shapiro_result = shapiro(data)
    print(f"Test Shapiro-Wilk pour la catégorie {category}: Statistique W = {shapiro_result.statistic}, P-value = {shapiro_result.pvalue}")

    if shapiro_result.pvalue <= 0.05:
        print("Les données ne suivent pas une distribution normale.\n")
    else:
        print("Les données suivent une distribution normale.\n")


# Réalisation du test de Levene pour tester l'égalité des variances
levene_result = levene(data_19_31, data_32_51, data_52_94)

print("Test de Levene - Statistique :", levene_result.statistic)
print("Test de Levene - P-value :", levene_result.pvalue)

if levene_result.pvalue <= 0.05:
    print("Les variances ne sont pas égales.\n")
else:
    print("Les variances sont égales.\n")
    

# le test de Kruskal-Wallis
from scipy.stats import kruskal

# Effectuer le test de Kruskal-Wallis
statistic, p_value = kruskal(data_19_31, data_32_51, data_52_94)
print("Test de Kruskal-Wallis:")
print("Statistique de test :", statistic)
print("p-value :", p_value)
print(p_value < alpha)

print("La p-value est inferieure au niveau de signification de 0.05, \nCela suggère qu'il y a suffisamment de preuves pour rejeter l'hypothèse nulle. \nJe peux conclure qu'il y a un lien entre l'age des clients et le montant du panier moyen.")
```

### Est-ce que les femmes achetent en moyenne pareil que les hommes  
H0 : panier moyen femmes = panier moyen des hommes

HA : Les femmes achetent differement des hommes 

```python
sns.catplot(x='sex', y='price', data=tcp, kind='box', showmeans=True)
plt.show()

alpha = 0.001

nouveau_df = tcp.groupby(['sex', 'client_id', 'session_id'])['price'].sum().reset_index()
nouveau_df_genre = nouveau_df.groupby(['client_id', 'sex'])['price'].mean().reset_index()

# je divise le DataFrame en deux groupes : hommes et femmes
mask = nouveau_df_genre['sex'] == 'f'
femme_prices = nouveau_df_genre.loc[mask, 'price'].reset_index()
homme_prices = nouveau_df_genre.loc[~mask, 'price'].reset_index()
achats_hommes = homme_prices['price']
achats_femmes = femme_prices['price']
    

from scipy.stats import kstest
# pour les data frame de plus de 5000 enregistrements
# Effectuer le test de Kolmogorov-Smirnov pour tester la normalité d'un grand echantillon > 5000
statistic, p_value = kstest(achats_femmes, 'norm')

print("Test de Kolmogorov-Smirnov pour tester la normalité de la distribution des achats femmes :")
print("Statistique de test femmes:", statistic)
if p_value < alpha:
    print("p-value :", p_value, "\nLes données ne suivent pas une distribution normale\n")
else:
    print("p-value :", p_value, "\nLes données suivent une distribution normale\n")
    
    
statistic, p_value = kstest(achats_hommes, 'norm')

print("Test de Kolmogorov-Smirnov pour tester la normalité de la distribution des achats hommes :")
print("Statistique de test hommes:", statistic)
if p_value < alpha:
    print("p-value :", p_value, "\nLes données ne suivent pas une distribution normale\n")
else:
    print("p-value :", p_value, "\nLes données suivent une distribution normale\n")


# Le test de Levene de la variance 
statistic, p_value = levene(homme_prices['price'], femme_prices['price'])
print("Résultats du test de Levene :")
print(f"Statistique de test : {statistic}")
print(f"P-value : {p_value}")

if p_value < 0.05:
    print("Les variances ne sont pas égales.\n")
else:
    print("Les variances sont égales.\n")

print("Les variances sont égales, j'ai un grand échantillon je vais utiliser le test Student car je suppose la normalité")


# je souppose la normalité car j'ai un grand échantillon et je réalise le test Student car mes deux échantillons sont equivalents et les variances proches
print('Test t de Student')
t_stat_achats, p_value_achats = stats.ttest_ind(achats_hommes, achats_femmes)

print("Résultats du test t de Student  :")
print(f"T-statistique : {t_stat_achats}")
print(f"P-value : {p_value_achats}")
print(p_value_achats<alpha, "\nIl n'y a pas suffisamment de preuves pour rejeter l'hypothèse nulle. Je ne peux pas conclure de manière \nsignificative qu'il y a une différence entre les moyennes d'achats des hommes et femmes.")
```
